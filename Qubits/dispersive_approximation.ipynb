# Qubit-Resonator Coupling

Rabi Hamiltonian, which is a qubit with frequency $\omega_q$, a resonator (along z axis) with quantized electric field $\hat{a}$. With electric dipole interaction between qubit and the electric field characterized by $g=d_x \mathcal{E}_0$:

$$
H_{\mathrm{Rabi}}=\omega_c\left(\hat{a}^{\dagger} \hat{a}+\frac{1}{2}\right)-\frac{1}{2} \omega_q \sigma_z-g\left(\hat{a}+\hat{a}^{\dagger}\right)\left(\sigma_{-}+\sigma_{+}\right)
$$

This Hamiltonian is represented in the direct products of bare states of cavity and qubits $\{|g\rangle|n\rangle,|e\rangle|n\rangle\}$, with corresponding bare states eigenvalues $\left\{ \pm \omega_q / 2+\omega_c(n+1 / 2)\right\}$.

## Jaynes-Cumming Approximation

Simplify the interaction Hamiltonian by rotating wave approximation (RWA), which is valid when both:
- $g \ll \omega_q, \omega_c$
- $\left|\omega_c-\omega_q\right| \ll\left|\omega_c+\omega_q\right|$

RWA in this context can also be interpreted as single photon regime. Then we reach the Jaynes-Cumming Hamiltonian:

$$
H_{\mathrm{JC}}=\omega_c\left(\hat{a}^{\dagger} \hat{a}+\frac{1}{2}\right)-\frac{1}{2} \omega_q \sigma_z-g\left(\hat{a}^{\dagger} \sigma_{-}+\hat{a} \sigma_{+}\right) .
$$

In the bare basis, it can be diamgonalized in infinite dimension:

$$
H_{\mathrm{J C}}=\left(\begin{array}{cccccc}
\frac{1}{2} \omega_c-\frac{\omega_q}{2} & 0 & 0 & 0 & 0 & 0 \\
0 & \frac{3}{2} \omega_c-\frac{\omega_q}{2} & g & 0 & 0 & 0 \\
0 & g & \frac{3}{2} \omega_c+\frac{\omega_q}{2} & 0 & 0 & 0 \\
& & & ... & & \\
0 & 0 & 0 & 0 & \left(n+\frac{1}{2}\right) \omega_c-\frac{\omega_q}{2} & \sqrt{n+1} g \\
0 & 0 & 0 & 0 & \sqrt{n+1} g & \left(n+\frac{1}{2}\right) \omega_c+\frac{\omega_q}{2}
\end{array}\right)=\left(
    \begin{array}{ccccc}
        \frac{1}{2} \omega_c-\frac{\omega_q}{2} & 0 & 0 & 0 & 0 \\
        0 & M_1 & 0 & 0 & 0 \\
        0 & 0 & M_2 & 0 & 0 \\
        & & ... & & \\
        0 & 0 & 0 & 0 & M_n
    \end{array}
\right)
$$

where $n \rightarrow \infty$, $\frac{1}{2} \omega_c-\frac{\omega_q}{2}$ is just describing the ground state $|g\rangle|0\rangle$, and each block $M_n$ is defined as:

$$
M_n=\left(\begin{array}{cc}
\left(n+\frac{1}{2}\right) \omega_c-\frac{\omega_q}{2} & \sqrt{n+1} g \\
\sqrt{n+1} g & \left(n+\frac{1}{2}\right) \omega_c+\frac{\omega_q}{2}
\end{array}\right)
$$

The eigenstates of $M_n$ together with $|g\rangle|0\rangle$ form a complete set of eigenstates for the entire qubit-cavity system. The eigenvalues of $H_\mathrm{JC}$ are:

$$
\begin{aligned}
E_g & =-\frac{\Delta}{2} \equiv -\frac{\omega_q-\omega_c}{2}\\
E_{\mp} & =(n+1) \omega_c \mp \frac{1}{2} \sqrt{4 g^2(n+1)+\Delta^2}
\end{aligned}
$$

The corresponding eigenstates are called the dressed states:

$$
\begin{aligned}
|0,-\rangle & =|g\rangle|0\rangle \\
|n,-\rangle & =\cos \left(\theta_n\right)|g\rangle|n+1\rangle-\sin \left(\theta_n\right)|e\rangle|n\rangle \\
|n,+\rangle & =\sin \left(\theta_n\right)|g\rangle|n+1\rangle+\cos \left(\theta_n\right)|e\rangle|n\rangle
\end{aligned}
$$

where $\theta_n$ is the "level of hybridization" (between the bare cavity and bare qubit states): 

$$
\theta_n=\frac{1}{2} \arctan(2 g \sqrt{n+1} / \Delta)
$$

## Dispersive Approximation

On top of thne Jaynes-Cumming's apprxoimation, when we have the cavity and the qubit far detuned:

- $g \ll \omega_q, \omega_c$
- $\left|\omega_c-\omega_q\right| \ll\left|\omega_c+\omega_q\right|$
- $|\Delta| \gg g$

we can further simplify the Hamiltonian. Note that this is the case in QPD, where $\omega_c \approx 8~\mathrm{GHz}$ and $\omega_q\approx 3~\mathrm{GHz}$

Introduce unitary transformation:

$$
\hat{T}=e^{\lambda\left(\sigma_{-} a^{\dagger}-\sigma_{+} a\right)}
$$

where $\lambda\equiv\frac{g}{\Delta}$. Apply this transformation to the JC Hamiltonian and evaluate all terms up to order $\lambda^2$, by taking advantage of the BCH relation $e^{\lambda \hat{B}} \hat{A} e^{-\lambda \hat{B}}=\hat{A}+\lambda[\hat{B}, \hat{A}]+\frac{\lambda^2}{2!}[\hat{B},[\hat{B}, \hat{A}]]+\mathcal{O}\left[\lambda^3\right]$:

$$
\begin{align*}
\hat{T} \hat{H}_\mathrm{J C} \hat{T}^{\dagger}&=\omega_c\left(\hat{a}^{\dagger} \hat{a}+\frac{1}{2}\right)-\frac{1}{2} \omega_q \sigma_z-\frac{g^2}{\Delta} \hat{a}^{\dagger} \hat{a} \sigma_z+\frac{g^2}{2 \Delta} \sigma_z \\
&\equiv \hat{H}_\mathrm{dis} + \frac{1}{2}\omega_c
\end{align*}
$$

The dispersive Hamiltonian describes the situation were the cavity and qubit are  far detuned and coupling is weak and dressed states are almost overlapping with the bare states. However, there is a very small interaction described by $-\frac{g^2}{\Delta} \hat{a}^{\dagger} \hat{a} \sigma_z$.

We can rearrange the terms:

$$
\hat{H}_{\mathrm{dis}}=\left(\omega_c-\chi \sigma_z\right) \hat{a}^{\dagger} \hat{a}-\frac{1}{2} \omega_{q\prime} \sigma_z
$$

where $\chi\equiv\frac{g^2}{\Delta}$ is defined as the dispersive shift or dispersive coupling rate, and $\omega_{q\prime} \equiv \omega_q + \frac{g^2}{\Delta}$ absorbs the "Lamb shift" into the bare qubit frequency $\omega_q$. 

The term $\chi$ can be also interpreted as "Lamb shift". As with the Lamb Shift in atomic orbitals, photons in the readout resonator, even if not resonant with the qubit transitions, can be virtually absorbed and reemitted by the qubit, causing a shift in the energy.

Note that this is already diagonalized in the bare basis! We see that the dispersive interaction is manifested as a qubit-state-dependent frequency shift for the cavity: the cavity frequency shifts by $\pm\chi$ when the qubit is in the ground or excited state.
