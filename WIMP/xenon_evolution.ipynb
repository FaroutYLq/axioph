This notebook shows the relative gain of each generation of XENON detectors. Below we put the limit on a 50 GeV DM with SI interaction.

Oct 24, 2024

```python
# Experiments performed already
# https://arxiv.org/pdf/0706.0039
year_cdms2  = 2006
sigma_cdms2 = 1.8e-43 
# https://arxiv.org/pdf/0706.0039
year_xe10   = 2007
sigma_xe10  = 5.0e-44 
# https://arxiv.org/pdf/1609.06154
year_xe100  = 2012
sigma_xe100 = 2.0e-45 
# https://arxiv.org/pdf/1805.12562
year_xe1t   = 2018
sigma_xe1t  = 8.0e-47 
# https://arxiv.org/pdf/2303.14729
year_xent   = 2022
sigma_xent  = 2.0e-47 

# Sensitivity projection
# https://arxiv.org/pdf/2007.08796
year_xentp  = 2029
sigma_xentp = 2.0e-48 
# https://arxiv.org/pdf/2404.19524 
year_xlzd   = 2040
sigma_xlzd  = 3.0e-49 # Assumed that XLZD will just reach atomspheric neutrino fog
```

```python
# Plotting style
import matplotlib.pyplot as plt
import matplotlib as mpl

mpl.rcParams["figure.dpi"] = 200
mpl.rcParams["figure.figsize"] = [4, 3]
mpl.rcParams["font.family"] = "serif"
mpl.rcParams["font.size"] = 9
mpl.rcParams["mathtext.fontset"] = "dejavuserif"
```

```python
plt.scatter(year_cdms2, sigma_cdms2, color="black")
plt.text(year_cdms2+1, sigma_cdms2, "CDMS", color="black", fontsize=7)

plt.scatter(year_xe10, sigma_xe10, color="tab:blue", label="Published")
plt.text(year_xe10+1, sigma_xe10, "XENON10(~$0.1M)", color="tab:blue", fontsize=7)

plt.scatter(year_xe100, sigma_xe100, color="tab:blue")
plt.text(year_xe100+1, sigma_xe100, "XENON100(~$0.5M)", color="tab:blue", fontsize=7)

plt.scatter(year_xe1t, sigma_xe1t, color="tab:blue")
plt.text(year_xe1t+1, sigma_xe1t, "XENON1T(~$10M)", color="tab:blue", fontsize=7)

plt.scatter(year_xent, sigma_xent, color="tab:blue")
plt.text(year_xent+1, sigma_xent, "XENONnT(~$30M) First Result", color="tab:blue", fontsize=7)

plt.scatter(year_xentp, sigma_xentp, color="tab:blue", facecolors='none', label="Projected")
plt.text(year_xentp-5, sigma_xentp*2, "XENONnT(~$30M) Projection", color="tab:blue", fontsize=7)

plt.scatter(year_xlzd, sigma_xlzd, color="tab:blue", facecolors='none')
plt.text(year_xlzd-13.5, sigma_xlzd*2, "XLZD(~$300M) Projection", color="tab:blue", fontsize=7)

plt.axhline(sigma_xlzd, color="grey", ls="--")

plt.yscale("log")
plt.legend()
plt.xlim(2005, 2043)
plt.ylim(bottom=0.5e-49)
plt.text(2018, 1e-49, "Neutrino Fog", color="grey")
plt.xlabel("Publication Year")
plt.ylabel("Sensitivity on 50 GeV WIMP $\sigma_{\mathrm{SI}}$ [cm$^2$]")
plt.tight_layout()
plt.savefig("xenon_si_limit_evolution.pdf")
```
