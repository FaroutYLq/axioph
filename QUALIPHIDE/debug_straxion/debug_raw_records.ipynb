```python
import straxion
```

```python
# Apply the style
plt.style.use('../../xenon.mplstyle')
```

```python
st = straxion.qualiphide()
```

```python
rr = st.get_array(
    "timeS429", # run id
    "raw_records", # data type
    config=dict(daq_input_dir="timeS429", record_length=5_000_000, fs=500_000)
)
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.plot(rr[0]['time'] + np.arange(rr[0]['length']) * rr[0]['dt'], 
         rr[0]['data_i'], label="I")
plt.plot(rr[0]['time'] + np.arange(rr[0]['length']) * rr[0]['dt'], 
         rr[0]['data_q'], label="Q")
plt.xlabel("Time Since Unix Epoch [ns]")
plt.ylabel("IQ [A.U.]")
plt.legend()
```
