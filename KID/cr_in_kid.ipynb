This notebook lays out the calculation toward estimation of a cosmic ray's signature in KID. Following Osmond Wen's thesis.

For small perturbations in $\delta n_{\mathrm{qp}}$, the $\delta S_{21}$ response is linear. 

$$
\delta S_{21}=\frac{Q_r^2}{Q_c} \alpha \gamma\left(\kappa_1+i \kappa_2\right) \delta n_{\mathrm{qp}}
$$

where we have assumed no off-resonance transmission and a real $Q_c$. We also assumed taylor expansion at initial $Q_i$ and $f_r$ to linear order, with $f=f_r$ all the time. This can be a wrong assumption because in reality you always measure the fixed freuqnecy $f$. This need to be corrected later for non-linearity.

- $1/Q_r = 1/Q_c + 1/Q_i$ is the total readout quality factor.
- $Q_c$ is the coupling quality factor.
- $\alpha \equiv \frac{L_s}{L_{\text {geom }}+L_s}$ is the fraction of kinetic inductance of the total inductance.
- $\kappa_1 \equiv \frac{d \sigma_1 /\left|\sigma_0\right|}{d n_{\mathrm{qp}}}=\frac{1}{\pi N_0} \sqrt{\frac{2}{\pi k T \Delta}} \sinh (x) K_0(x)$ with $x=\hbar \omega / 2 k T$ is the response of real conductance to change in quasiparticle. 
    - $\sigma_0$ is the zero-temperature conductance.
- $\kappa_2 \equiv \frac{d \sigma_2 /\left|\sigma_0\right|}{d n_{\mathrm{qp}}}=-\frac{1}{2 N_0 \Delta}\left(1+\sqrt{\frac{2 \Delta}{\pi k T}} e^{-x} I_0(x)\right)$ is the response of imaginary conductance to change in quasiparticle. 
- $\delta n_\mathrm{qp} = \frac{\eta_\mathrm{pb}}{\Delta V_\mathrm{SC}}$ is the change in quasiparticle numbers
    - $\eta_\mathrm{p b}=\frac{\Delta}{h\langle\nu\rangle} \frac{\langle\delta x\rangle}{\left(\delta x / \delta N_\mathrm{q p}\right)}=0.32$ from Peter's PRX paper.

Considering the fact that resonant frequency and $Q_i$ changes when the perturbation is big enough, one has to insert

$$
\begin{gathered}
\frac{1}{Q_r}=\frac{1}{Q_{r, 0}}+\delta \frac{1}{Q_i} \\
f_r=f_{r, 0}+\delta f_r \Rightarrow \frac{f-f_r}{f_r}=\frac{-\delta f_r}{f_{r, 0}+\delta f_r} .
\end{gathered}
$$

into 

$$
S_{21}(f)=1-\frac{Q_r}{Q_c} \frac{1}{1+2 i Q_r \frac{f-f_r}{f_r}}
$$

to get the full expression for 

$$
\delta S_{21}=S_{21}\left(f_r\right)-S_{21}\left(f_{r, 0}\right)
$$

where "$_0$" denotes the initial parameters when there is no perturbation.

Some algebra:

$$
\delta S_{21}=1-\frac{Q_r}{Q_c} \frac{1}{1+2 i Q_r \frac{f-f_r}{f_r}}-\left(1-\frac{Q_{r, 0}}{Q_c}\right)
$$

Assuming $\delta f_r \ll f_{r, 0}$, one gets

# Visualization

Assuming linearity, one have this in the resonator coordinate

$$
\begin{align}
    \delta S_{21}
    &\approx\frac{Q_r^2}{Q_c} \alpha \gamma\left(\kappa_1+i \kappa_2\right) \delta n_{\mathrm{qp}} \\
    &\approx\frac{Q_r^2}{Q_c} \alpha \gamma\left(\frac{1}{\pi N_0} \sqrt{\frac{2}{\pi k T \Delta}} \sinh (x) K_0(x) - i \frac{1}{2 N_0 \Delta}\left(1+\sqrt{\frac{2 \Delta}{\pi k T}} e^{-x} I_0(x)\right)\right) \frac{\eta_\mathrm{pb} E_\mathrm{abs}}{V_\mathrm{sc} \Delta}
\end{align}
$$

Considering non-linear effects:

$$
\begin{align}
    \delta S_{21} &\approx \frac{Q_{r, 0}}{Q_c}\left(1-\frac{1}{1+Q_{r, 0} \alpha \gamma\left(\delta \sigma_1+i \delta \sigma_2\right) /\left|\sigma_0\right|}\right) \\
    &= \frac{Q_{r, 0}}{Q_c}\left(1-\frac{1}{1+Q_{r, 0} \alpha \gamma\left(\frac{2 \Delta_0}{\hbar \omega} \frac{1}{N_0 \sqrt{2 \pi k T \Delta_0}} \sinh (\xi) K_0(\xi) \sigma_n \cdot \delta n_{\mathrm{qp}} -i\frac{\pi \Delta_0}{\hbar \omega} \frac{1}{2 N_0 \Delta_0}\left(1+\sqrt{\frac{2 \Delta_0}{\pi k T}} e^{-\xi} I_0(\xi)\right) \sigma_n \cdot \delta n_{\mathrm{qp}}\right) /\left|\sigma_0\right|}\right)
\end{align}
$$

Few assumptions and details:
- $\Delta=\Delta_0$, as the low-temperature limit. For the exact temperature dependence, see Gao Eqn. 2.88.
- $x\equiv\hbar \omega / 2 k T$ is the temperature-reduced photon energy.
- $1/Q_r \approx 1/Q_{r,0} \equiv 1/Q_{i,0} + 1/Q_{c,0}$.
- $\alpha=\alpha_0$.
- $K_0$ and $I_0$ are the 0-th order modified Bessel function of the first and second kind.

For each ideal resonator, we always have 

$$
S_{21}(f; f_r)=1-\frac{Q_r}{Q_c} \frac{1}{1+2 i Q_r \frac{f-f_r}{f_r}}
$$

```python
import matplotlib.pyplot as plt
import numpy as np

# Apply the style
plt.style.use('../xenonnt.mplstyle')

def S21(f, fr, Qi, Qc):
    """
    Calculate the S21 transmission parameter for a coupled resonator.
    
    Parameters:
    -----------
    f : float or array-like
        Frequency (Hz)
    fr : float
        Resonance frequency (Hz)
    Qi : float
        Resonator intrinsic quality factor
    Qc : float
        Coupling quality factor
    
    Returns:
    --------
    complex or array of complex
        S21 transmission coefficient
    
    Formula:
    --------
    S21(f; fr) = 1 - (Qr/Qc) * 1/(1 + 2i*Qr*(f-fr)/fr)
    """
    # Convert inputs to numpy arrays for vectorized operations
    f = np.asarray(f)
    
    # Calculate the complex denominator
    Qr = 1/(1/Qi + 1/Qc)
    denominator = 1 + 2j * Qr * (f - fr) / fr
    
    # Calculate S21
    s21 = 1 - (Qr / Qc) * (1 / denominator)
    
    return s21
```

```python
def delta_fr(fr0, alpha, gamma, delta_sigma2, sigma0):
    """
    Calculate the resonance frequency shift due to changes in conductivity.
    
    Parameters:
    -----------
    fr0 : float or array-like
        Zero-temperature resonance frequency (Hz)
    alpha : float
        Fraction of kinetic inductance (0 <= alpha <= 1)
    gamma : float
        Penetration depth factor (-1, -1/2, or -1/3)
    delta_sigma2 : float or array-like
        Change in imaginary part of conductivity (S/m)
    sigma0 : complex
        Zero temperature conductivity (S/m)
    
    Returns:
    --------
    float or array
        Resonance frequency shift δfr (Hz)
    
    Formula:
    --------
    δfr/fr0 = -1/2 * α * γ * δσ2/|σ0|
    """
    # Convert inputs to numpy arrays for vectorized operations
    fr0 = np.asarray(fr0)
    delta_sigma2 = np.asarray(delta_sigma2)
    
    # Validate gamma value
    valid_gamma = [-1, -1/2, -1/3]
    if gamma not in valid_gamma:
        print(f"Warning: gamma = {gamma} is not a typical value.")
        print(f"Common values are: {valid_gamma}")
    
    # Validate alpha
    if not (0 <= alpha <= 1):
        print(f"Warning: alpha = {alpha} is outside typical range [0, 1]")
    
    # Calculate the magnitude of sigma0
    sigma0_magnitude = np.abs(sigma0)
    
    # Calculate the fractional frequency shift
    fractional_shift = -0.5 * alpha * gamma * (delta_sigma2 / sigma0_magnitude)
    
    # Calculate absolute frequency shift
    delta_fr_result = fractional_shift * fr0
    
    return delta_fr_result


from scipy.constants import hbar, pi

def sigma_0(Delta, omega, sigma_n=3.8e7):
    """
    Calculate the zero temperature conductivity for a superconductor.
    
    Parameters:
    -----------
    Delta : float or array-like
        Superconducting gap energy (J)
    omega : float or array-like
        Photon angular frequency (rad/s)
    sigma_n : float or complex
        Normal state conductivity (S/m), default to Aluminum (3.8e7 S/m).
    
    Returns:
    --------
    complex or array of complex
        Zero temperature conductivity σ₀ (S/m)
    
    Formula:
    --------
    σ₀ = -i * (π * Δ / ℏω) * σₙ
    
    Note:
    -----
    This formula applies in the limit ℏω >> Δ (high frequency limit)
    and represents the response of Cooper pairs in a superconductor.
    """
    # Convert inputs to numpy arrays for vectorized operations
    Delta = np.asarray(Delta)
    omega = np.asarray(omega)
    sigma_n = np.asarray(sigma_n)
    
    # Calculate sigma_0
    sigma_0_result = -1j * (pi * Delta / (hbar * omega)) * sigma_n
    
    return sigma_0_result

def sigma_0_eV(Delta_eV, f_Hz, sigma_n=3.8e7):
    """
    Calculate the zero temperature conductivity with convenient units.
    
    Parameters:
    -----------
    Delta_eV : float or array-like
        Superconducting gap energy in eV
    f_Hz : float or array-like
        Photon frequency in Hz
    sigma_n : float or complex
        Normal state conductivity (S/m), default to Aluminum (3.8e7 S/m).
    
    Returns:
    --------
    complex or array of complex
        Zero temperature conductivity σ₀ (S/m)
    """
    from scipy.constants import eV
    
    # Convert to SI units
    Delta_J = Delta_eV * eV  # Convert eV to Joules
    omega = 2 * pi * f_Hz    # Convert Hz to rad/s
    
    return sigma_0(Delta_J, omega, sigma_n)
```

```python
from scipy.constants import hbar, k, pi, eV
from scipy.special import i0, k0  # Modified Bessel functions I_0 and K_0

def calculate_xi(omega, T):
    """
    Calculate the dimensionless parameter ξ = ℏω/(2kT).
    
    Parameters:
    -----------
    omega : float or array-like
        Angular frequency (rad/s)
    T : float or array-like
        Temperature (K)
    
    Returns:
    --------
    float or array
        Dimensionless parameter ξ
    """
    return hbar * omega / (2 * k * T)

def calculate_xi_hz(f_Hz, T):
    """
    Calculate the dimensionless parameter ξ = ℏω/(2kT).
    """
    return hbar * 2 * pi * f_Hz / (2 * k * T)

def delta_sigma1(Delta0, omega, T, N0, sigma_n, delta_n_qp):
    """
    Calculate the change in real part of conductivity δσ₁.
    
    Parameters:
    -----------
    Delta0 : float
        Superconducting gap at T=0 (J)
    omega : float or array-like
        Angular frequency (rad/s)
    T : float or array-like
        Temperature (K)
    N0 : float
        Density of states at Fermi level (states/eV/um³)
    sigma_n : float
        Normal state conductivity (S/m)
    delta_n_qp : float or array-like
        Change in quasiparticle density (um⁻³)
    
    Returns:
    --------
    float or array
        Change in real conductivity δσ₁ (S/m)
    
    Formula:
    --------
    δσ₁ = (2Δ₀/ℏω) × (1/(N₀√(2πkTΔ₀))) × sinh(ξ)K₀(ξ) × σₙ × δn_qp
    """
    # Convert to numpy arrays
    omega = np.asarray(omega)
    T = np.asarray(T)
    delta_n_qp = np.asarray(delta_n_qp)
    
    # Calculate xi
    xi = calculate_xi(omega, T)
    
    # Calculate the prefactor
    prefactor = (2 * Delta0) / (hbar * omega)
    normalization = 1 / (N0 / eV * np.sqrt(2 * pi * k * T * Delta0))
    
    # Calculate Bessel function terms
    sinh_xi = np.sinh(xi)
    K0_xi = k0(xi)  # Modified Bessel function K_0
    
    # Full formula
    delta_sigma1_result = prefactor * normalization * sinh_xi * K0_xi * sigma_n * delta_n_qp
    
    return delta_sigma1_result

def delta_sigma2(Delta0, omega, T, N0, sigma_n, delta_n_qp):
    """
    Calculate the change in imaginary part of conductivity δσ₂.
    
    Parameters:
    -----------
    Delta0 : float
        Superconducting gap at T=0 (J)
    omega : float or array-like
        Angular frequency (rad/s)
    T : float or array-like
        Temperature (K)
    N0 : float
        Density of states at Fermi level (states/eV/um³)
    sigma_n : float
        Normal state conductivity (S/m)
    delta_n_qp : float or array-like
        Change in quasiparticle density (um⁻³)
    
    Returns:
    --------
    float or array
        Change in imaginary conductivity δσ₂ (S/m)
    
    Formula:
    --------
    δσ₂ = -(πΔ₀/ℏω) × (1/(2N₀Δ₀)) × (1 + √(2Δ₀/πkT) × e^(-ξ)I₀(ξ)) × σₙ × δn_qp
    """
    # Convert to numpy arrays
    omega = np.asarray(omega)
    T = np.asarray(T)
    delta_n_qp = np.asarray(delta_n_qp)
    
    # Calculate xi
    xi = calculate_xi(omega, T)
    
    # Calculate the prefactor
    prefactor = -(pi * Delta0) / (hbar * omega)
    normalization = 1 / (2 * N0 / eV * Delta0)
    
    # Calculate the correction term
    sqrt_term = np.sqrt(2 * Delta0 / (pi * k * T))
    exp_bessel_term = np.exp(-xi) * i0(xi)  # Modified Bessel function I_0
    correction = 1 + sqrt_term * exp_bessel_term
    
    # Full formula
    delta_sigma2_result = prefactor * normalization * correction * sigma_n * delta_n_qp
    
    return delta_sigma2_result

def delta_sigma1_eV(delta_n_qp, Delta0_eV, f_Hz, T, N0, sigma_n):
    """
    Calculate δσ₁ with gap energy in eV and frequency in Hz.
    
    Parameters:
    -----------
    Delta0_eV : float
        Superconducting gap at T=0 (eV)
    f_Hz : float or array-like
        Frequency (Hz)
    T : float or array-like
        Temperature (K)
    N0 : float
        Density of states at Fermi level (states/eV/m³)
    sigma_n : float
        Normal state conductivity (S/m)
    delta_n_qp : float or array-like
        Change in quasiparticle density (m⁻³)
    
    Returns:
    --------
    float or array
        Change in real conductivity δσ₁ (S/m)
    """
    from scipy.constants import eV
    
    Delta0_J = Delta0_eV * eV
    omega = 2 * pi * f_Hz
    
    return delta_sigma1(Delta0_J, omega, T, N0, sigma_n, delta_n_qp)

def delta_sigma2_eV(delta_n_qp, Delta0_eV, f_Hz, T, N0, sigma_n):
    """
    Calculate δσ₂ with gap energy in eV and frequency in Hz.
    
    Parameters:
    -----------
    Delta0_eV : float
        Superconducting gap at T=0 (eV)
    f_Hz : float or array-like
        Frequency (Hz)
    T : float or array-like
        Temperature (K)
    N0 : float
        Density of states at Fermi level (states/eV/m³)
    sigma_n : float
        Normal state conductivity (S/m)
    delta_n_qp : float or array-like
        Change in quasiparticle density (m⁻³)
    
    Returns:
    --------
    float or array
        Change in imaginary conductivity δσ₂ (S/m)
    """
    from scipy.constants import eV
    
    Delta0_J = Delta0_eV * eV
    omega = 2 * pi * f_Hz
    
    return delta_sigma2(Delta0_J, omega, T, N0, sigma_n, delta_n_qp)

def delta_n_qp_eV(Eabs, Vsc, Delta0_eV, eta_pb=0.01, eta_phonon=1):
    """Compute the change in quasiparticle density.
    """
    return (Eabs / (Vsc * Delta0_eV)) * eta_pb * eta_phonon
```

```python
def delta_S21_explicit(Eabs, Vsc=18, eta_pb=0.2, eta_phonon=1, Delta0_eV=0.00020, f_Hz=4.039e9, T=20e-3, N0=1.72e10, sigma_n=3.8e7, Qi0=334381, Qc=17958, alpha=0.04, gamma=-1):
    """Compute the change in S21 with non-linear effects.
    """
    Qr0 = 1/(1/Qi0 + 1/Qc)
    delta_n_qp = delta_n_qp_eV(Eabs, Vsc, Delta0_eV, eta_pb, eta_phonon)
    delta_sigma1 = delta_sigma1_eV(delta_n_qp, Delta0_eV, f_Hz, T, N0, sigma_n)
    delta_sigma2 = delta_sigma2_eV(delta_n_qp, Delta0_eV, f_Hz, T, N0, sigma_n)
    sigma_0 = sigma_0_eV(Delta0_eV, f_Hz, sigma_n)

    # Not sure why but seems abs(gamma) is needed here
    #delta_s21 = Qr0/Qc * (1-1/(1+ Qr0 * alpha * np.abs(gamma) * (delta_sigma1 + 1j * delta_sigma2) / np.abs(sigma_0)))
    # Not sure why but seems minus sign is needed here instead of plus. It is the same problem as the previous negative kappa1 thing
    delta_s21 = Qr0/Qc * (1-1/(1+ Qr0 * (
        alpha * gamma * (-delta_sigma1) / np.abs(sigma_0) -
        1j * 2 / (1 - 1/(0.5 * alpha * gamma * delta_sigma2 / np.abs(sigma_0)))
    )))
    return delta_s21
```

```python
def fit_circle_complex(complex_array):
    """
    Fit a circle to an array of complex numbers in the complex plane.
    
    Parameters:
    -----------
    complex_array : array-like of complex numbers
        Array of complex numbers to fit a circle to
    
    Returns:
    --------
    dict
        Dictionary containing:
        - 'center': complex number representing circle center
        - 'radius': float representing circle radius
        - 'residuals': array of residuals for each point
        - 'rms_error': root mean square error of the fit
    
    Method:
    -------
    Uses algebraic circle fitting (Kasa method) which minimizes the sum of 
    squared algebraic distances from points to the circle.
    """
    # Convert to numpy array and extract real/imaginary parts
    z = np.asarray(complex_array)
    x = z.real
    y = z.imag
    
    # Number of points
    n = len(z)
    
    # Set up the linear system Ax = b for circle fitting
    # Circle equation: (x - xc)² + (y - yc)² = r²
    # Expanding: x² - 2*xc*x + xc² + y² - 2*yc*y + yc² = r²
    # Rearranging: 2*xc*x + 2*yc*y + (r² - xc² - yc²) = x² + y²
    
    # Matrix A: each row is [2*x, 2*y, 1]
    A = np.column_stack([2*x, 2*y, np.ones(n)])
    
    # Vector b: x² + y²
    b = x**2 + y**2
    
    # Solve the least squares problem
    try:
        # Use SVD for robust solution
        solution, residuals, rank, s = np.linalg.lstsq(A, b, rcond=None)
        
        # Extract circle parameters
        xc = solution[0]  # x-coordinate of center
        yc = solution[1]  # y-coordinate of center
        c = solution[2]   # r² - xc² - yc²
        
        # Calculate radius
        radius = np.sqrt(c + xc**2 + yc**2)
        
        # Calculate residuals (distance from each point to the fitted circle)
        distances = np.sqrt((x - xc)**2 + (y - yc)**2)
        point_residuals = np.abs(distances - radius)
        
        # RMS error
        rms_error = np.sqrt(np.mean(point_residuals**2))
        
        return {
            'center': complex(xc, yc),
            'radius': radius,
            'residuals': point_residuals,
            'rms_error': rms_error,
            'fit_quality': 'good' if rms_error < 0.1 * radius else 'poor'
        }
        
    except np.linalg.LinAlgError:
        raise ValueError("Unable to fit circle - points may be collinear or insufficient data")
```

In linear approximation ($\delta \rightarrow d$), from OW 6.36 and 6.37 (fixed the missing factor of $-\frac{1}{2}$)

$$
\begin{align}
    Q_i(T)&=\frac{1}{\frac{1}{Q_{i, 0}} + \alpha \gamma \kappa_1 \delta n_\mathrm{qp}}\\
    f_r &= f_{r, 0} \left(1 - \frac{1}{2}\alpha \gamma \kappa_2 \delta n_\mathrm{qp} \right)
\end{align}
$$

It is evident that $\frac{1}{Q_i}$ can change a lot, while $f_r$ seems fine. So I think the explicit calculation might be necessary.

$$
\begin{align}
    \frac{1}{Q_i} &= \frac{1}{Q_{i,0}} + \left(\alpha \gamma \kappa_1 - \frac{\alpha^2 R}{\omega L_s} \kappa_2 \right) \delta n_\mathrm{qp}
\end{align}
$$

```python
# Default values of kappa from OW Fig. 6.4. kappa2 is negative from Gao Fig. 2.10.
# But it seems positive kappa1 is more reasonable, otherwise you get negative Q_i.
def fr_linear(delta_n_qp, kappa2=-4e-7, fr0=4.039e9, alpha=0.04, gamma=-1):
    return fr0 * (1 - 0.5 * alpha * gamma * kappa2 * delta_n_qp)

    # An additional factor of 1/3 seem to match theoretical non-linear \delta S_{21} better, not understood yet
    #return fr0 * (1 - 0.3 * 0.5 * alpha * gamma * kappa2 * delta_n_qp)

def Qi_linear(delta_n_qp, kappa1=2e-7, Qi0=334381, alpha=0.04, gamma=-1):
    return 1 / (1 / Qi0 + alpha * np.abs(gamma) * kappa1 * delta_n_qp)
```

```python
def get_theta(Eabs, Vsc=18, Delta0_eV=0.00020, eta_pb=0.2, eta_phonon=1, Qc=17958, Qi0=334381, fr0=4.039e9):
    """Compute inverse angle in resonator coordinate"""
    delta_s21s = delta_S21_explicit(Eabs=Eabs, Vsc=Vsc, Delta0_eV=Delta0_eV, eta_pb=eta_pb, eta_phonon=eta_phonon)
    origin = S21(f=fr0, fr=fr0, Qi=Qi0, Qc=Qc)
    center = (1+origin.real)/2 + 0 * 1j
    s21s = origin + delta_s21s
    thetas = -(np.angle(s21s - center) - np.pi)
    return thetas
```
